```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm  import SVC
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

iris-dataset.png

Data

```python
iris= datasets.load_iris()
iris
```

```python
X=iris.data[:, :2]
y=iris.target
```

```python

# Plotting
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
X_plot = np.c_[xx.ravel(), yy.ravel()]
```

# **ML**

## Linear Kernel

```python
# linear Kernel:
svc_linear = svm.SVC(kernel= "linear", C= 5.0)
svc_linear.fit(X,y)

# z: y_pred
z = svc_linear.predict(X_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (8,5))
plt.contour(xx, yy,z, alpha = 0.3)
plt.scatter(x[:,0], x[:,1], c=y)
plt.xlabel("Sepal lenth")
plt.ylabel("Sepal width")
plt.title("SVM with Linear Kernel")
```

## Polynomial Kernel

```python
# Polynomial Kernel:
svm_poly = svm.SVC(kernel= "poly", C= 5.0)
svm_poly.fit(X,y)

# z: y_pred
z = svm_poly.predict(X_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (15,5))
plt.subplot(121)
plt.contour(xx, yy,z, alpha = 0.3)
plt.set_cmap("gist_rainbow")
plt.scatter(X[:,0], X[:,1], c=y, cmap = plt.cm.Set1)
plt.xlabel("Sepal length")
plt.xlim(xx.min(), xx.max())
plt.ylabel("Sepal width")
plt.title("SVM with polynomial Kernel")
plt.plot()
```

## Sigmoidal kernel

```python
#sigmoidal  Kernel:
svm_sigmoid = svm.SVC(kernel= "sigmoid", C= 1.0)
svm_sigmoid.fit(X,y)

# z: y_pred
z = svm_sigmoid.predict(X_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (8,5))
plt.contour(xx, yy,z, alpha = 0.3)
plt.scatter(x[:,0], x[:,1], c=y)
plt.xlabel("Sepal lenth")
plt.ylabel("Sepal width")
plt.title("SVM with Sigmoid Kernel")
```

## RBF Kernel

```python
# RBF Kernel:
svm_rbf = svm.SVC(kernel= "rbf", C= 5)
svm_rbf.fit(X,y)

# z: y_pred
z = svm_rbf.predict(X_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (15,5))
plt.subplot(121)
plt.set_cmap("gist_rainbow")
plt.contour(xx, yy,z, alpha = 0.3)
plt.scatter(X[:,0], X[:,1], c=y, cmap = plt.cm.Set1)
plt.xlabel("Sepal lenth")
plt.xlim(xx.min(), xx.max())
plt.ylabel("Sepal width")
plt.title("SVM with RBF Kernel")
plt.plot()
```
